```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from collections import Counter
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
blob_df = pd.read_csv("blob.csv")
dart_df = pd.read_csv("dart.csv")
outliers_df = pd.read_csv("outliers.csv")
spiral2_df = pd.read_csv("spiral2.csv")
basic2_df = pd.read_csv("basic2.csv")
boxes3_df = pd.read_csv("boxes3.csv")
```

```python
pd.set_option('display.max_rows', None)       # Display all rows
pd.set_option('display.max_columns', None)    # Display all columns
pd.set_option('display.width', None)          # No limit on width
pd.set_option('display.max_colwidth', None)
```

```python
df_blob_kmeans = blob_df.copy()
df_dart_kmeans = dart_df.copy()
df_basic_kmeans = basic2_df.copy()
df_outliers_kmeans = outliers_df.copy()
df_spiral2_kmeans = spiral2_df.copy()
df_boxes3_kmeans = boxes3_df.copy()
```

```python
df_blob_kmeans.drop(['color'], axis = 1, inplace =True)
df_dart_kmeans.drop(['color'], axis = 1, inplace =True)
df_basic_kmeans.drop(['color'], axis = 1, inplace =True)
df_outliers_kmeans.drop(['color'], axis = 1, inplace =True)
df_spiral2_kmeans.drop(['color'], axis = 1, inplace =True)
df_boxes3_kmeans.drop(['color'], axis = 1, inplace =True)
```

```python
kmeans_blob = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
kmeans_dart = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
kmeans_basic = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
kmeans_outliers = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
kmeans_spiral2 = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
kmeans_boxes3 = KMeans(n_clusters = 12, init = 'k-means++', random_state = 42)

y_kmeans_blob = kmeans_blob.fit_predict(df_blob_kmeans)
y_kmeans_dart = kmeans_dart.fit_predict(df_dart_kmeans)
y_kmeans_basic = kmeans_basic.fit_predict(df_basic_kmeans)
y_kmeans_outliers = kmeans_outliers.fit_predict(df_outliers_kmeans)
y_kmeans_spiral2 = kmeans_spiral2.fit_predict(df_spiral2_kmeans)
y_kmeans_boxes3 = kmeans_boxes3.fit_predict(df_boxes3_kmeans)
```

```python
# Creating 'Cluster' columns in data sets
df_blob_kmeans['Cluster'] = y_kmeans_blob
df_dart_kmeans['Cluster'] = y_kmeans_dart
df_basic_kmeans['Cluster'] = y_kmeans_basic
df_outliers_kmeans['Cluster'] = y_kmeans_outliers
df_spiral2_kmeans['Cluster'] = y_kmeans_spiral2
df_boxes3_kmeans['Cluster'] = y_kmeans_boxes3
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(13, 13))
#fig.suptitle('EXERCISE DATA SETS\n', size=18)

axes[0,0].scatter(blob_df['x'], blob_df['y'], c=blob_df['color'], s=10, cmap="viridis")
axes[0,0].set_title("Blob")

axes[0,1].scatter(dart_df['x'], dart_df['y'], c=dart_df['color'], s=10, cmap="plasma")
axes[0,1].set_title("Dart")

axes[0,2].scatter(basic2_df['x'], basic2_df['y'], c=basic2_df['color'], s=10, cmap="cool")
axes[0,2].set_title("Basic")

axes[1,0].scatter(outliers_df['x'], outliers_df['y'], c=outliers_df['color'], s=10, cmap="spring")
axes[1,0].set_title("Outliers")

axes[1,1].scatter(spiral2_df['x'], spiral2_df['y'], c=spiral2_df['color'], s=10, cmap="autumn")
axes[1,1].set_title("Spiral")

axes[1,2].scatter(boxes3_df['x'], boxes3_df['y'], c=boxes3_df['color'], s=10, cmap = "Set3")
axes[1,2].set_title("Boxes");

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Adjusted figsize for better scaling of all plots
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(8, 24))  # Increased nrows to 6 for all the plots
fig.subplots_adjust(top=0.9, hspace=0.4)  # Adjust space at the top and between rows for clarity
#fig.suptitle('ANSWER vs K-Means\n', size=18)

# Row 1: Blob
axes[0, 0].scatter(blob_df['x'], blob_df['y'], c=blob_df['color'], s=10, cmap="viridis")
axes[0, 0].set_title("Blob")
axes[0, 1].scatter(df_blob_kmeans['x'], df_blob_kmeans['y'], c=df_blob_kmeans['Cluster'], s=10, cmap="viridis")
axes[0, 1].set_title("K-Means Blob")

# Row 2: Dart
axes[1, 0].scatter(dart_df['x'], dart_df['y'], c=dart_df['color'], s=10, cmap="plasma")
axes[1, 0].set_title("Dart")
axes[1, 1].scatter(df_dart_kmeans['x'], df_dart_kmeans['y'], c=df_dart_kmeans['Cluster'], s=10, cmap="plasma")
axes[1, 1].set_title("K-Means Dart")

# Row 3: Basic
axes[2, 0].scatter(basic2_df['x'], basic2_df['y'], c=basic2_df['color'], s=10, cmap="cool")
axes[2, 0].set_title("Basic")
axes[2, 1].scatter(df_basic_kmeans['x'], df_basic_kmeans['y'], c=df_basic_kmeans['Cluster'], s=10, cmap="cool")
axes[2, 1].set_title("K-Means Basic")

# Row 4: Outliers
axes[3, 0].scatter(outliers_df['x'], outliers_df['y'], c=outliers_df['color'], s=10, cmap="spring")
axes[3, 0].set_title("Outliers")
axes[3, 1].scatter(df_outliers_kmeans['x'], df_outliers_kmeans['y'], c=df_outliers_kmeans['Cluster'], s=10, cmap="spring")
axes[3, 1].set_title("K-Means Outliers")

# Row 5: Spiral
axes[4, 0].scatter(spiral2_df['x'], spiral2_df['y'], c=spiral2_df['color'], s=10, cmap="autumn")
axes[4, 0].set_title("Spiral")
axes[4, 1].scatter(df_spiral2_kmeans['x'], df_spiral2_kmeans['y'], c=df_spiral2_kmeans['Cluster'], s=10, cmap="autumn")
axes[4, 1].set_title("K-Means Spiral")

# Row 6: Boxes
axes[5, 0].scatter(boxes3_df['x'], boxes3_df['y'], c=boxes3_df['color'], s=10, cmap="Set3")
axes[5, 0].set_title("Boxes")
axes[5, 1].scatter(df_boxes3_kmeans['x'], df_boxes3_kmeans['y'], c=df_boxes3_kmeans['Cluster'], s=10, cmap="Set3")
axes[5, 1].set_title("K-Means Boxes")

# Adjust layout to make sure the figure fits well
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to leave space for the title
plt.show()


```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

blob_silhouette_score = silhouette_score(df_blob_kmeans, y_kmeans_blob)
dart_silhouette_score = silhouette_score(df_dart_kmeans, y_kmeans_dart)
basic_silhouette_score = silhouette_score(df_basic_kmeans, y_kmeans_basic)
outliers_silhouette_score = silhouette_score(df_outliers_kmeans, y_kmeans_outliers)
spiral2_silhouette_score = silhouette_score(df_spiral2_kmeans, y_kmeans_spiral2)
boxes3_silhouette_score = silhouette_score(df_boxes3_kmeans, y_kmeans_boxes3)

# Print silhouette scores for all datasets
print("Silhouette Scores:")
print(f"Blob: {blob_silhouette_score}")
print(f"Dart: {dart_silhouette_score}")
print(f"Basic: {basic_silhouette_score}")
print(f"Outliers: {outliers_silhouette_score}")
print(f"Spiral: {spiral2_silhouette_score}")
print(f"Boxes: {boxes3_silhouette_score}")
```

```python

import matplotlib.pyplot as plt

# Silhouette scores for each dataset
datasets = ['Blob', 'Dart', 'Basic', 'Outliers', 'Spiral', 'Boxes']
silhouette_scores = [blob_silhouette_score, dart_silhouette_score, basic_silhouette_score, 
                     outliers_silhouette_score, spiral2_silhouette_score, boxes3_silhouette_score]

# Colors for each bar
colors = ['skyblue', 'lightgreen', 'salmon', 'lightcoral', 'lightpink', 'lightyellow']

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(datasets, silhouette_scores, color=colors, edgecolor='black')

# Add silhouette score values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.01, round(yval, 2), ha='center', va='bottom', fontsize=12)

# Set labels and title
plt.xlabel('Dataset', fontsize=14)
plt.ylabel('Silhouette Score', fontsize=14)
plt.title('Silhouette Scores for K-Means Clustering on Different Datasets', fontsize=16)
plt.ylim([0, 1])  # Silhouette score range is from -1 to 1, but typically [0, 1] is more meaningful
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()

```

```python
from sklearn.metrics import adjusted_rand_score

# Extract the true labels from the 'Cluster' column
y_true_blob = blob_df['color'].values  # True labels for the Blob dataset
y_true_dart =dart_df['color'].values  # True labels for the Dart dataset
y_true_basic = basic2_df['color'].values  # True labels for the Basic dataset
y_true_outliers =outliers_df['color'].values  # True labels for the Outliers dataset
y_true_spiral2 =spiral2_df['color'].values  # True labels for the Spiral dataset
y_true_boxes3 = boxes3_df['color'].values  # True labels for the Boxes dataset

# Calculate ARI for all datasets
blob_ari = adjusted_rand_score(y_true_blob, y_kmeans_blob)
dart_ari = adjusted_rand_score(y_true_dart, y_kmeans_dart)
basic_ari = adjusted_rand_score(y_true_basic, y_kmeans_basic)
outliers_ari = adjusted_rand_score(y_true_outliers, y_kmeans_outliers)
spiral2_ari = adjusted_rand_score(y_true_spiral2, y_kmeans_spiral2)
boxes3_ari = adjusted_rand_score(y_true_boxes3, y_kmeans_boxes3)

# Print ARI for all datasets
print("Adjusted Rand Index Scores:")
print(f"Blob: {blob_ari}")
print(f"Dart: {dart_ari}")
print(f"Basic: {basic_ari}")
print(f"Outliers: {outliers_ari}")
print(f"Spiral: {spiral2_ari}")
print(f"Boxes: {boxes3_ari}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ARI scores for all datasets
ari_scores = {
    "Blob": blob_ari,
    "Dart": dart_ari,
    "Basic": basic_ari,
    "Outliers": outliers_ari,
    "Spiral": spiral2_ari,
    "Boxes": boxes3_ari
}

# Create a DataFrame for better visualization
ari_df = pd.DataFrame(list(ari_scores.items()), columns=["Dataset", "Adjusted Rand Index"])

# Sort the ARI scores in descending order
ari_df = ari_df.sort_values(by="Adjusted Rand Index", ascending=False)

# Set up the plot
plt.figure(figsize=(12, 8))
ax = sns.barplot(x="Adjusted Rand Index", y="Dataset", data=ari_df, palette="viridis")

# Add title and labels
#plt.title("Adjusted Rand Index Scores for DBSCAN on Different Datasets")
plt.xlabel("Adjusted Rand Index")
plt.ylabel("Dataset")

# Annotate the bars with the ARI values
for i, v in enumerate(ari_df["Adjusted Rand Index"]):
    ax.text(v + 0.005, i, f"{v:.3f}", va="center", color="black")

# Show plot
plt.show()

```
